```python
import xml.etree.ElementTree as ET
from collections import defaultdict
rows = defaultdict(list)
text = response.text

root = ET.fromstring(response.text)


for node in root.iter('item'):
    for i in node:
        print(i.tag, i.text)
        rows[i.tag] = i.text
print(rows)

    
```

```python
import pandas as pd
rows_d = dict(rows)
sd = pd.DataFrame(rows_d, columns=rows_d.keys(), index=[0])
sd.to_csv("sample_2010_01.csv", mode='w')
```
