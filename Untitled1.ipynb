```python
import pandas as pd
```

```python
df=pd.read_csv('final_weather_2004_2011_8.csv', encoding='utf-8')
one_day = df[df['tm']=='2011-01-01 00:00']
```

```python
sigun = pd.read_csv('2018년소멸위험지수.csv', encoding='cp949')
sigun['sigun_nm']=sigun['sigun_nm'].map(lambda x:x[:-1])
sigun_small=sigun[['sigun_nm','sigun_code']]
```

```python
sigun_small
```

```python
one_day_sample=pd.merge(left = one_day , right = sigun_small, how = "inner", left_on = "stnNm", right_on = "sigun_nm")
one_day_sample
```

```python
one_day_sample.to_csv('one_day.csv')
```
